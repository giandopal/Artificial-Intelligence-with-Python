```python
import geopandas as gpd
from shapely.geometry import Point

# Crea un dataframe con le coordinate dei punti
lon = [10.19, -4.97, 139.68, -73.94]
lat = [56.46, 39.92, 35.69, 40.73]
geometry = [Point(x, y) for x, y in zip(lon, lat)]
df = gpd.GeoDataFrame({'geometry': geometry})

# Crea la figura e i suoi assi
fig, ax = plt.subplots(figsize=(18, 18))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.plot(ax=ax, color='lightgray', edgecolor='white')

# Plotta i punti
df.plot(ax=ax, markersize=30, color='red', marker='o')

# Mostra la figura
plt.show()
```
