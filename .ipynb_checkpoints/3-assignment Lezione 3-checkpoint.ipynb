7. Plotta la funzione $y=x^3 - 2x^2 + x -2$ tra -3 e 3 utilizzando la vettorizzazione delle funzioni

8. Plotta la funzione $y=-2x^3 + x^2 - x +1$ tra -3 e 3 utilizzando la vettorizzazione delle funzioni

Vedi la soluzione sotto


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3,3,100)
def myfunction(x):
    y = x**3-3*x**2+x-2
    return y
def myfunction2(x):
    y = -2*x**3+x**2-x+1
    return y
myfunc_vect = np.vectorize(myfunction)
myfunc_vect2 = np.vectorize(myfunction2)
y = myfunc_vect(x)
y2=myfunc_vect2(x)
plt.plot(x,y)
plt.plot(x,y2)
```
